```python
df = pd.DataFrame([], columns=['Заголовок', 'Ссылки', 'Дата/Время', 'Раздел', 'Просмотры', 'Количество комментариев'])# создаем датафрейм, добавляем пустой список и колонки с теми названиями, с которыми мы хотим
j = 0 # создаем переменные для циклов
l = 0
t = 0
w = 0
for p in range(1, 10): # создаем цикл, перебираем страницы на сайте
  print(p) 
  url = f'https://www.e1.ru/text/?rubric=politics&page={p}' 
  r = requests.get(url,cookies=cookies,headers=headers) # создаем запросы на нужный нам раздел сайта, добавляем данные с html(network) Copy as cUrl(bash), конвертируем код в https://curlconverter.com, чтобы сайт не считывал как взлом
  soup = bs4.BeautifulSoup(r.text) # создаем красивый суп) преобразуем HTML-код в сложное дерево объектов, таких как тег, строка или комментарий.
  items = soup.findAll('h2', {'class': 'h9Jmx'}) # находим небходимые нам теги, в которых  текстовые данные 
  time = soup.findAll('time', {'class':'_2DfZq'}) # находим небходимые нам теги, в которых  временные данные+дата
  see = soup.findAll('div',{'class':'e0SBr'})# находим небходимые нам теги, в которых данные по просмотрам сайта, хотя это массив, но там хранятся нужные нам данные
  comment = soup.findAll('a',{'class':'+N8hV l4RQV'}) # находим небходимые нам теги, в которых данные по комментариям



  for i in time: # создаем цикл, перебираем время новостей начиная с актуальных
    df.loc[l, 'Дата/Время'] = i('a')[0].text # обращаемся с помощью loc к нужной нам колонке в которой будет информация про время+дату и с помощью тега а вытаскиваем время и дату в текстовом виде, указывая обязательно индекс, как бы начало отсчета
    l +=1 # перебираем время и дату
  
  for s in see:  # создаем цикл, перебираем просмотры новостей начиная с актуальных
    df.loc[t, 'Просмотры'] = s('span')[0].text # обращаемся с помощью loc к нужной нам колонке в которой будет информация про просмотры, находим нужный нам тег span и обращаемся к текстовым данным, указывая индекс
    t +=1 # перебираем просмотры

  for c in comment: # создаем цикл, перебираем время новостей начиная с актуальных
    df.loc[w, 'Количество комментариев'] = c('span')[0].text # обращаемся с помощью loc к нужной нам колонке в которой будет информация про время+дату и с помощью тега а вытаскиваем время и дату в текстовом виде, указывая обязательно индекс, как бы начало отсчета
    w +=1
  
  for item in items: # создаем цикл, перебираем заголовки, ссылки и разделы новостей начиная с актуальных
    title = item.find('a').get_text(strip=True) # находим заголовки в тегах и вытаскиваем текст
    df.loc[j, 'Заголовок'] = title  # обращаемся с помощью loc к нужной нам колонке в которой будет информация про заголовки 
    df.loc[j, 'Ссылки'] = 'https://www.e1.ru' + item.find('a').get('href')  # обращаемся с помощью loc к нужной нам колонке в которой будет информация про ссылки, чтобы создать ссылку необходима ссылка начала сайта+ он будет находить в теге а конец ссылки новостей
    df.loc[j, 'Раздел'] = 'Политика'  # обращаемся с помощью loc к нужной нам колонке в которой будет информация про раздел, в данном случае эта политика 
    j += 1 # перебираем заголовки, ссылки и разделы
```

```python
df[['Дата', 'Время']] = df['Дата/Время'].str.split(',', expand=True) # разделяем столбец Дата/Время на Дату и Время
```
