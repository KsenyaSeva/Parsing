```python
df = pd.DataFrame([], columns=['Заголовок', 'Ссылки', 'Дата/Время', 'Раздел', 'Просмотры', 'Количество комментариев'])
j = 0 
l = 0
t = 0
w = 0
for p in range(1, 10): 
  print(p) 
  url = f'https://www.e1.ru/text/?rubric=politics&page={p}' 
  r = requests.get(url,cookies=cookies,headers=headers) 
  soup = bs4.BeautifulSoup(r.text)
  items = soup.findAll('h2', {'class': 'h9Jmx'}) 
  time = soup.findAll('time', {'class':'_2DfZq'}) 
  see = soup.findAll('div',{'class':'e0SBr'})
  comment = soup.findAll('a',{'class':'+N8hV l4RQV'}) 


  for i in time: 
    df.loc[l, 'Дата/Время'] = i('a')[0].text 
    l +=1 
  
  for s in see:  
    df.loc[t, 'Просмотры'] = s('span')[0].text 
    t +=1

  for c in comment: 
    df.loc[w, 'Количество комментариев'] = c('span')[0].text 
    w +=1
  
  for item in items: 
    title = item.find('a').get_text(strip=True) 
    df.loc[j, 'Заголовок'] = title   
    df.loc[j, 'Ссылки'] = 'https://www.e1.ru' + item.find('a').get('href')  
    df.loc[j, 'Раздел'] = 'Политика'  
    j += 1 
```

```python
df[['Дата', 'Время']] = df['Дата/Время'].str.split(',', expand=True) 
```
